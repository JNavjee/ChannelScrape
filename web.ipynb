```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import NoSuchElementException
import time
import csv
import requests
import datetime
import pandas as pd
from bs4 import BeautifulSoup
```

```python
service = Service('C:\\Users\\206823200\\OneDrive - NBCUniversal\\chromedriver-win64\\chromedriver-win64\\chromedriver.exe')

chrome_options = webdriver.ChromeOptions()

driver = webdriver.Chrome(service=service, options =chrome_options)

chrome_options = Options()
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")  # Prevents memory issues
chrome_options.add_argument("--disable-gpu")  # Disable GPU rendering for stability
chrome_options.add_argument("--remote-debugging-port=9222")  # Enable remote debugging
chrome_options.add_argument("--headless=new")  # Runs Chrome in headless mode for stability
chrome_options.add_argument("--disable-software-rasterizer")

def accept_cookies():
    try:
        accept_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, 'button.fc-button.fc-cta-consent.fc-primary-button'))
        )
        accept_button.click()
        print("Accepted cookies.")
    except NoSuchElementException:
        print("No cookie banner found.")
    except Exception as e:
        print(f"Error accepting cookies: {e}")

driver.get('https://www.citizen.digital/tv/citizen-tv-15/schedules')

WebDriverWait(driver, 100).until(
    EC.presence_of_element_located((By.CLASS_NAME, 'tv-progslist'))
)

accept_cookies()

WebDriverWait(driver, 100)

soup = BeautifulSoup(driver.page_source, "html.parser")

schedule_items = soup.find_all('div', class_='row')

print(f"Found {len(schedule_items)} schedule items.")

times = []
titles = []

for index, item in enumerate(schedule_items):
    time_container = item.find('div', class_='col-md-2')
    if time_container:
        time_element = time_container.find('span')
        if time_element:
            time = time_element.get_text(strip=True)
            times.append(time)
        else:
            print(f"Time element not found for item {index}.")
    else:
        print(f"Time container not found for item {index}.")
    
    title_container = item.find('div', class_='vidpg-video-excerpt')
    if title_container:
        title_element = title_container.find('h3')
        if title_element:
            title = title_element.get_text(strip=True)
            titles.append(title)
        else:
            print(f"Title element not found for item {index}.")
    else:
        print(f"Title container not found for item {index}.")

print("Times:", times)
print("Titles:", titles)

df = pd.DataFrame({
    'Title': titles,
    'Time': times
})

print(df)

df.to_csv('CitizenTV2.csv', index=False)
```
